```python
from vidstab import VidStab
import matplotlib.pyplot as plt

stabilizer = VidStab()
stabilizer.stabilize(input_path='Vid/sine.mov', output_path='Output/sine.avi')
```

```python
stabilizer.plot_trajectory()
plot_trajectory = plt.show()
```

```python
stabilizer.plot_transforms()
plot_transform = plt.show()
```

```python
import pandas as pd
# Convert transforms to a DataFrame and export to Excel
transforms_df = pd.DataFrame(stabilizer.transforms, columns=['dx', 'dy', 'da'])
transforms_df.to_excel('Output/sine.xlsx', index=False)
```
